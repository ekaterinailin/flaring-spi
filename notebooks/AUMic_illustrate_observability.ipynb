```python
from funcs.notebook import *
import matplotlib.pyplot as plt

path = "/home/ekaterina/Documents/002_writing/aumic-flaring-spi-draft/figures"
```

```python
x = np.linspace(-.5,.5,1000)
y = np.zeros_like(x) 
```

```python
y[250:750] = 0.5 * np.cos((x[250:750])*2*np.pi)
```

```python
fig, ax = plt.subplots(figsize=(8,6))
plt.plot(x,y,c="k",linewidth=2)
plt.xlim(-.5,.5)
plt.ylim(0,1.3)
centers = [-.26,0.,.24]
plt.scatter(centers,[1.,1.,1.], s=2000, facecolor="k", edgecolor="k")
plt.scatter([-.29,0.,.27],[1.04,1.04,1.04], s=50, facecolor="#EE5717", edgecolor="k")
for c in centers:
    plt.axvline(c,ymin=.82,ymax=.85,c="k",linewidth=3)
plt.axhline(0.4,linestyle="dashed", c="k")
ax.fill_between(x,y1=0.,y2=y,facecolor="#578DB6",  where=(y>0.),hatch=".")
ax.fill_between(x,y1=0.4,y2=y, facecolor="#0C4A79", where=(y>0.4), hatch=".")

ax.axes.yaxis.set_ticks([0.])
plt.xlabel("region phase")
plt.ylabel("number of flares")
plt.text(s="detection threshold", x=0.14, y=0.42)
plt.text(s="flaring\nregion", x=-.41, y=1.1, color="#EE5717")
plt.text(s="equator-on\nstar", x=.28, y=1.1, color="k")
plt.text(s="observable\nflares", x=-.1, y=.55, color="#0C4A79")
plt.text(s="visible\nflares", x=-.35, y=.1, color="#578DB6")
ax.scatter(centers,[1.15]*3, s=300, marker=r'$\circlearrowleft$',c="k",zorder=100)
plt.legend(frameon=False,)
plt.tight_layout()
plt.savefig(f"{path}/observability_illustration.png",dpi=300);
```

### caption

Number of flares produced by a single flaring region on the stellar surface at different phases of the region. The flares become visible (grey area) when the region comes into view. However, they are not observable because the amplitude is geometrically foreshortened when the region is close to the limb. When the region is close enough to the center of the stellar disk, the flares' amplitude exceeds the detection threshold so that they can be observed (blue area). The phase can be the stellar rotation phase, the orbital phase of the planet, or the synodic phase of the star-planet system.
