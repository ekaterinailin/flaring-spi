# YZ Cet



```python
offset = 2452996 
t0 = 2452996.25 - offset
omega = 197 / 180 * np.pi
P = 2.02087
e = 0.06
```

```python
def rseq(tperi, e=0., P=1., t0=0, omega=0):
    """set to 0 and solve for tperi"""
    M = 2 * np.pi * (t0 - tperi) / P# % (2 * np.pi)
    return M + (2 * e - .25 * e**3) * np.sin(M) + 1.25 * e**2 * np.sin(2 * M) + 13 / 12 * e**3 * np.sin(3 * M) + omega
    
```

```python
tperi = np.linspace(-4, 4, 100)
plt.plot(tperi, rseq(tperi, e=e, P=P, t0=t0, omega=omega))
plt.axhline(0)
```

```python
tperi_initial_guess = 1
g = lambda x: rseq(x, e=e, P=P, t0=t0, omega=omega)
tperi_solution = fsolve(g, tperi_initial_guess)
```

```python
tperi_solution[0] + offset
```
