```python
from altaipony.fakeflares import aflare

path = "/home/ekaterina/Documents/002_writing/flaring-spi-draft/flaring-spi-draft/figures"
```

```python
time=np.linspace(0,1,22)
np.random.seed(54)
flux = aflare(time, .12,0.065,.76) + np.random.normal(0,.01,22)
```

```python
plt.figure(figsize=(8,5.5))
plt.plot(time, flux, c="grey", zorder=-10)
plt.scatter(time[:1], flux[:1],c="k",marker=".",label="quiescent flux")
plt.scatter(time[1:3], flux[1:3],c="r",marker="x",s=70,)
plt.scatter(time[3:7], flux[3:7],c="k",marker="*",s=70,label="flare signal above 3$\sigma$ threshold")
plt.scatter(time[7:11], flux[7:11],c="r",marker="x",s=70,label="padded signal before and after flare")
plt.scatter(time[11:], flux[11:],c="k",marker=".")
plt.axhline(0.03, zorder=-20, linestyle="dotted",label=r"$3\sigma$ noise threshold")
plt.axhline(0.0, zorder=-20,linestyle="dashed",)
plt.xlabel("time")
plt.ylabel("relative flux above median")
plt.xlim(0,1)
plt.legend(frameon=False)
plt.tight_layout()
plt.savefig(f"{path}/sigma_clipping.png", dpi=300)
```
