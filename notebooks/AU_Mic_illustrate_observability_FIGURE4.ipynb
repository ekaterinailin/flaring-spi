```python
from funcs.notebook import *
import matplotlib.pyplot as plt

path = "/home/ekaterina/Documents/002_writing/aumic-flaring-spi-draft/figures"
```

```python
# create a set of phases
x = np.linspace(-.5,.5,1000)

# no SPI flares seen 
y = np.zeros_like(x) 

# add SPI flares with a cosine
y[250:750] = 0.5 * np.cos((x[250:750])*2*np.pi)
```

```python
# set up figure
fig, ax = plt.subplots(figsize=(8,6))

# plot SPI flare number
plt.plot(x, y ,c="k",linewidth=2)

# draw little spotted stars

# center star
centers = [-.26, 0., .24]

#  plot stellar surface
plt.scatter(centers, [1., 1., 1.], s=2000, facecolor="k", edgecolor="k")
plt.text(s="equator-on\nstar", x=.28, y=1.1, color="k")

# plot spots
plt.scatter([-.29, 0., .27], [1.04, 1.04, 1.04], s=50, facecolor="#EE5717", edgecolor="k")
plt.text(s="flaring\nregion", x=-.41, y=1.1, color="#EE5717")

# draw rotation axis
ax.scatter(centers,[1.15]*3, s=300, marker=r'$\circlearrowleft$',c="k",zorder=100)
for c in centers:
    plt.axvline(c, ymin=.82, ymax=.85, c="k", linewidth=3)

# draw detection threshold
plt.axhline(0.4,linestyle="dashed", c="k")
plt.text(s="detection threshold", x=0.14, y=0.42)

# fill area where flares are visible
ax.fill_between(x,y1=0.,y2=y,facecolor="#578DB6",  where=(y>0.),hatch=".")
plt.text(s="visible\nflares", x=-.35, y=.1, color="#578DB6")

# draw where flares are observable
ax.fill_between(x,y1=0.4,y2=y, facecolor="#0C4A79", where=(y>0.4), hatch=".")
plt.text(s="observable\nflares", x=-.1, y=.55, color="#0C4A79")

# remove all ticks on y axis except the zero
ax.axes.yaxis.set_ticks([0.])

# layout
plt.xlim(-.5,.5)
plt.ylim(0,1.3)
plt.xlabel("region phase")
plt.ylabel("number of flares")
plt.tight_layout()
# plt.savefig(f"{path}/observability_illustration.png",dpi=300);
```

### caption:

Number of flares produced by a single flaring region on the stellar surface at different phases of the region. The flares become visible (dark blue area) when the region comes into view. However, they are not observable because the amplitude is geometrically foreshortened when the region is close to the limb. When the region is close enough to the center of the stellar disk, the flares' amplitude exceeds the detection threshold so that they can be observed (light blue area). The phase can be the stellar rotation phase, the orbital phase of the planet, or the synodic phase of the star-planet system.
